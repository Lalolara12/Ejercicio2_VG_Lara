## Tablas

### Tablas de frecuencias
### Tablas de contingencia
Mediante las tablas de contingencia se clasifica un conjunto de observaciones en función de los valores de dos variables cualitativas que definen, respectivamente, las filas y columnas de
la tabla. Una tabla MN es la que tiene M filas y N columnas, y las celdas pueden representar frecuencias, porcentajes de fila, porcentajes de columna o porcentajes del total de observaciones.

### Medidas de dispersión
Varianza -  cuantifica el grado de variabilidad de los datos en torno a la media
Desviación típica - la raíz cuadrada positiva de la varianza -  Esta medida da idea de la dispersión de los datos con respecto a su media aritmética
Coeficiente de variación - es una medida de variabilidad relativa que relaciona la desviación estándar de un conjunto de observaciones con su media

## Medidas de 
Recorrido -  mide la amplitud de las observaciones y se calcula como la diferencia entre los valores máximo y mínimo
Recorrido intercuartílico - se calcula como la diferencia entre el tercer y el primer cuartil corresponde con el recorrido de los datos que ocupan el 50% central de las
observaciones.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data_df = pd.DataFrame(iris.data, columns= iris.feature_names)
data_df["target"] = iris.target

data_df
```

```python
# Tomar ciertas columnas
columns = ["sepal length (cm)", "petal length (cm)", "target"]
data_df = data_df[columns]

data_df = data_df[data_df["target"].isin([0,1,])]

data_df.tail(10)
```

```python
plt.figure()

# datos de iris setosa
setosa_data = data_df[data_df["target"] == 0]

# datos de iris versicolor
versicolor_data = data_df[data_df["target"] == 1]

# datos de iris virginica
virginica_data = data_df[data_df["target"] == 2]

# graficar los datos 
#setosa
plt.scatter(setosa_data["sepal length (cm)"], setosa_data["petal length (cm)"], label="setosa", color="red", s=20)
# versicolor
plt.scatter(versicolor_data["sepal length (cm)"], versicolor_data["petal length (cm)"], label="versicolor", color="magenta", s=20)
# virginica
plt.scatter(virginica_data["sepal length (cm)"], virginica_data["petal length (cm)"], label="virginica", color="cyan", s=20)

# personalizar el grafico
plt.title('sepal length vs petal length')
plt.xlabel('Sepal length (cm)')
plt.ylabel('Petal length (cm)')
plt.grid()
plt.legend()

# mostrar el grafico
plt.show()
```

```python
from sklearn.datasets import fetch_openml

titanic = fetch_openml ('titanic', version= 1, as_frame= True)
titanic_frame = pd.DataFrame(titanic.data)
titanic_frame
```

```python
conteo_hombres = titanic_frame[titanic_frame["sex"] == "male"]["sex"].count()
conteo_mujeres = titanic_frame[titanic_frame["sex"] == "female"]["sex"].count()

conteo_general = [conteo_mujeres, conteo_hombres]

genero = titanic_frame["sex"].unique()

print(f"Hombres: {conteo_hombres}, Mujeres: {conteo_mujeres}")
```

```python
fig, ax = plt.subplots(1, 2)

# Gráfico de barras verticales
ax[0].bar(genero, conteo_general)
ax[0].set_title('Barra Vertical')

ax[1].barh(genero, conteo_general)
ax[1].set_title('Barra Horizontal')
plt.show()
```

```python
# suma de los valores de cada clase
conteo_primeraClase = titanic_frame[titanic_frame["pclass"] == 1]["pclass"].count()
conteo_segundaClase = titanic_frame[titanic_frame["pclass"] == 2]["pclass"].count()
conteo_terceraClase = titanic_frame[titanic_frame["pclass"] == 3]["pclass"].count()
conteo_total = [conteo_primeraClase, conteo_segundaClase, conteo_terceraClase]

etiquetas = ["Primera Clase", "Segunda Clase", "Tercera Clase"]

colores = ["red", "yellow", "green"]

extraccion = 0.1,0,0

print(f"primera: {conteo_primeraClase}, segunda: {conteo_segundaClase}, tercera: {conteo_terceraClase}")
plt.figure()

plt.title("Porcentaje de pasajeros en cada clase", fontsize=15, fontweight= "bold")

plt.pie(conteo_total, labels = etiquetas, 
        autopct="%0.1f%%", 
        colors = colores, 
        explode = extraccion, 
        shadow = True, 
        startangle = 90)
plt.show()
```

```python
df_histograma = titanic_frame.dropna(subset=['age'], axis= 'index')
 
# crear histograma

df_histograma.hist(
    'age',
    by = 'sex',
    bins = 10
)

plt.show()
```

```python

df_histograma = pd.DataFrame(titanic_frame)
df_cajas = df_histograma
df_cajas.boxplot(column='age')

# Mostrar el gráfico
plt.title('Boxplot de Edad')
plt.ylabel('Edad')
plt.show()
```

```python
grouped_data = [df_cajas[df_cajas['sex'] == gender]['age'].dropna() for gender in df_cajas['sex'].unique()]

# Crear el diagrama de caja
plt.boxplot(grouped_data, labels=df_cajas['sex'].unique())

# Añadir etiquetas y título
plt.xlabel('Sexo')
plt.ylabel('Edad')
plt.title('Diagrama de caja de la edad por sexo')

# Mostrar el gráfico
plt.show()
```
